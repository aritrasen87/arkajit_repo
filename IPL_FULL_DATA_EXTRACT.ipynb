```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
import numpy as np
```

```python
m_url = r'http://www.iplt20.com/stats/2017'
m_2017 = urllib.request.urlopen(m_url)
m_html_2017 = BeautifulSoup(m_2017)
m_rows = m_html_2017.find_all('tr')
```

```python
m_data = []
head =[row.find_all('th') for row in m_rows]
m_data.append([ele.text.strip() for ele in head[0]])
for row in m_rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    m_data.append([ele for ele in cols if ele !=[]])
```

```python
m_data
```

```python
df_m_2017 = pd.DataFrame(m_data[2:], columns = m_data[0])
```

```python
df_m_2017
```
