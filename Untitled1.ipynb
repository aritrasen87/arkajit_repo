```python
def load_data(start , end ):
    matches = []
    for i in range(start, end):
        if os.path.isfile('./'+str(i)+'.yaml'):
            with open('./'+str(i)+'.yaml','r')as f:
                doc= yaml.load(f)
                matches.append(doc)
    return matches
    
```
